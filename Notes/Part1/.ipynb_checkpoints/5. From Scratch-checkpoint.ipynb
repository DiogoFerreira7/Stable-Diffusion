```python
from torch import tensor

# we turn the column into the tensors and this is the dependent data, the one we are trying to predict
t_dep = tensor(df.Survived)
```

```python
indep_cols = ['Age', 'SibSp', 'Parch', 'LogFare']

t_indep = tensor(df[indep_cols].values, dtype = torch.float)
t_indep
```

```python
t_indep.shape
```

```python
len(t_indep.shape)
```

```python
# by generating a manual seed we will make sure that our accuracy is not being affected by randomness
torch.manual_seed(442)

n_coeff = t_indep.shape[1]
coeffs = torch.rand(n_coeff) - 0.5 # centered at a mean of 0 and between 0.5 and -0.5
coeffs
```

```python
# this broadcasting is done in optimised GPUs and in C code - all the fast code is being ran on very quick hardware
# as long as the last axis matches we can broadcast anything with anything else
t_indep * coeffs
```

```python
# the problem with adding up the data is that the data needs to have aroudn similar distributions
preds = (t_indep * coeffs).sum(axis = 1)
preds
```

```python
loss = torch.abs(preds - t_dep).mean()
loss
```

```python
def calc_preds(coeffs, indeps): return (indeps * coeffs).sum(axis = 1)
def calc_loss(coeffs, indeps, deps): return torch.abs(calc_preds(coeffs, indeps) - deps).mean()
```

```python
# We now have all of the basic steps and we can carry out a gradient descent step
coeffs.requires_grad_() # _ at the end is an inplace operation that will chagne the actual values
```

```python
loss = calc_loss(coeffs, t_indep, t_dep)
loss
```

```python
loss.backward()
# computes the gradients of the loss function with respect to the loss of the network
```

```python
coeffs.grad
# now that we have accessed the gradients that we previously told torch to require and did backwards on the gradients we have access to them
```

```python
with torch.no_grad():
    coeffs.sub_(coeffs.grad * 0.1) # subtract in place the gradient * the learning rate
    print(calc_loss(coeffs, t_indep, t_dep))
```
